**Auxiliares ejercicios 2, 3 y 4**

```python
def filter_matriz(A: np.array, condition):
    res = zerosLike(A)
    for i, row in enumerate(A):
        for j, value in enumerate(row):
            if condition(i, j):
                res[i][j] = value
    return res

def zerosLike(A: np.array):
    zerosA = A.copy()
    for a in zerosA:
        for i in range(len(a)):
            a[i] = 0
    return zerosA
```

**Ejercicio 5**

```python
def traza(A: np.array):
    res = 0
    for i, row in enumerate(A):
        for j, value in enumerate(row):
            if i == j:
                res += value
    return res

traza(a_matrix)
```

**Ejercicio 6**

```python
def traspuesta(A: np.array):
    res = zerosLike(A)
    for i, row in enumerate(A):
        for j, value in enumerate(row):
            res[i][j] = A[j][i]
            
    return res

traspuesta(a_matrix)
```

```python
def esSimetrica(A: np.array):
    for i, row in enumerate(A):
        for j, value in enumerate(row):
            if A[j][i] != A[i][j]:
                return False
    return True
    
esSimetrica(a_matrix)
```

```python
def calcularAx(A: np.array, x: np.array):
    res = np.zeros(A.shape[0])  #A.shape[0] devuelve la cantidad de filas de A
    for i, row in enumerate(A):
        for j, value in enumerate(row):
            res[i] += value * x[j] 
    return res

calcularAx(a_matrix, np.array([1,1,1,1]))
            
```

**Ejercicio 10**

```python
def sumar_fila_multiplo(A: np.array, i: int, j: int, s: int):
    jxEscalar = A[j].copy()
    multiplicarxEscalar(s, jxEscalar)
    sumarAVector(A[i], jxEscalar)

def multiplicarxEscalar(k: int, v: np.array):
    for i in range(len(v)):
        v[i] *= k

def sumarAVector(v: np.array, u: np.array):
    for i in range(len(v)):
        v[i] += u[i]

otromas_matrix = a_matrix.copy()
sumar_fila_multiplo(otromas_matrix, 0, 1, 2)
otromas_matrix
```

**Ejercicio 11**

```python
def esDiagonalmenteDominante(A: np.array):
    for i, row in enumerate(A):
        row_value = 0
        for j, value in enumerate(row):
            row_value += abs(value)
        diagonal_element = abs(A[i][i])
        if row_value - diagonal_element > diagonal_element:
            return False
    return True

diagonally_dominant_matrix = np.array([[ 312, 11, 14, 17], [ 4,  543,  6,  7], [ 8,  9, 348, 11], [12, 13, 14, 842]])
esDiagonalmenteDominante(diagonally_dominant_matrix)
```

**Ejercicio 12**

```python
def matrizCirculante(v: np.array):
    matrix = np.empty((len(v), len(v)))
    matrix[0] = v
    for i in range(1, len(v)):
        matrix[i] = siguiente_ciclo(matrix[i-1])
    return matrix

def siguiente_ciclo(v: np.array):
    res = np.empty(len(v))
    for i in range(1, len(v)):
        res[i] = v[i-1]
    res[0] = v[len(v)-1]
    return res

vector = np.array([1,2,3,4,5])
matrizCirculante(vector)        
```

**Ejercicio 13**

```python
def matrizVandermonde(v: np.array):
    res = np.empty((len(v), len(v)))
    for i in range (len(v)):
        res[i] = potencia_por_elemento(v, i)
    return res

def potencia_por_elemento(v: np.array, k: int):
    res = v.copy()
    for i, value in enumerate(res):
        res[i] = value ** k
    return res

matrizVandermonde(vector)    
```

**Ejercicio 14**


```python


def multiplicacion_matrices(A: np.array, M: np.array):
    res = np.empty((A.shape[0], M.shape[1]))
    n = A.shape[1]
    for i in range(len(A)):
        for j in range(M.shape[1]):
            value = 0
            for k in range(n):
                value += A[i][k] * M[k][j]
            res[i][j] = value
    return res  

def potencia_matriz(A:  np.array, k: int):
    res = A.copy()
    for _ in range(k - 1):
        res = multiplicacion_matrices(res, A)
    return res

mi_matriz = np.array([[2,2],[3,3]])
potencia_matriz(mi_matriz, 3)
    
```
